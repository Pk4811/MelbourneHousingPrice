### **Finding the Right Balance: Underfitting vs. Overfitting**  

At this point, you should have a good grasp of underfitting and overfitting—two common challenges in building accurate machine learning models. Now, let's explore how we can adjust our models to get the best possible predictions.  

### **Experimenting with Different Models**  

Now that you have a reliable way to measure model accuracy, it's time to experiment with different approaches to see which one works best. But what options do you have?  

Scikit-learn’s Decision Tree model, for example, comes with many customization options. One of the most important settings is **tree depth**, which determines how many times the data is split before reaching a final prediction.  

### **Understanding Tree Depth**  

Imagine a decision tree that only makes **two splits** before predicting a house price. This would be a very **shallow tree**, meaning it groups houses into broad categories, making its predictions less precise.  

On the other hand, if the tree keeps splitting until it reaches **10 levels deep**, it creates **1024 different groups (leaves)**. Each group contains only a handful of houses, so the model can make highly specific predictions for those training houses.  

However, this can lead to **overfitting**—when a model learns the training data too well but struggles to generalize to new data.  

---

### **Overfitting vs. Underfitting**  

- **Overfitting:** The model learns every tiny detail in the training data, leading to very accurate predictions on known data but poor performance on new data.  
- **Underfitting:** The model is too simple, failing to capture important patterns in the data, leading to poor accuracy even on training data.  

What we want is a **sweet spot**—a model complex enough to make accurate predictions but not so complex that it memorizes the data instead of learning patterns.  

---

### **Finding the Right Tree Depth**  

One way to control this balance is by using the **max_leaf_nodes** parameter in a Decision Tree.  
- Fewer leaves = **simpler model** (reduces overfitting).  
- More leaves = **more complex model** (reduces underfitting but risks overfitting).  

By testing different values for **max_leaf_nodes**, you can find the point where the model performs best on new, unseen data. This ensures you’re building a model that’s both **accurate and reliable**. 🚀

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python

# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))

# Write loop to find the ideal tree size from candidate_max_leaf_nodes
my_mae = {max_leaf_nodes: get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y) for max_leaf_nodes in [5, 50, 500, 5000]}
```

```python
# Store the best value of max_leaf_nodes (it will be either 5, 25, 50, 100, 250 or 500)
best_tree_size = min(my_mae, key=my_mae.get)
# Fill in argument to make optimal size and uncomment
final_model = DecisionTreeRegressor(max_leaf_nodes = best_tree_size, random_state=1)

# fit the final model and uncomment the next two lines
final_model.fit(X, y)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))
```

```python
# To improve accuracy, create a new Random Forest model which you will train on all training data
rf_model_on_full_data = RandomForestRegressor(random_state=1)

# fit rf_model_on_full_data on all data from the training data
rf_model_on_full_data.fit(X, y)
```
